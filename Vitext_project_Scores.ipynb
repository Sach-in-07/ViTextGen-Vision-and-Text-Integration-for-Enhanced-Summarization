```python
import requests
from bs4 import BeautifulSoup
import os
import hashlib
from PIL import Image
import io
import pandas as pd
from pathlib import Path
from urllib.parse import urljoin

# Function to get the content of the page using requests
def get_content_from_url(url):
    response = requests.get(url)
    return response.text

# Function to parse image URLs from the content
def parse_image_urls(content):
    soup = BeautifulSoup(content, "html.parser")
    results = []
    for img in soup.findAll("img"):
        img_src = img.get("src")
        if img_src and img_src not in results:
            results.append(img_src)
    return results

# Function to save image URLs to a CSV file
def save_urls_to_csv(image_urls, file_path="links.csv"):
    df = pd.DataFrame({"links": image_urls})
    df.to_csv(file_path, index=False, encoding="utf-8")
    print(f"Image URLs saved to {file_path}")

# Function to download and save images to a directory
def get_and_save_image_to_file(image_url, output_dir, base_url=None):
    # Ensure output directory exists
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    # Handle relative URLs by joining with base_url
    if not image_url.startswith(('http://', 'https://')):
        if base_url:
            image_url = urljoin(base_url, image_url)
        else:
            image_url = 'https://' + image_url.lstrip('/')

    try:
        image_content = requests.get(image_url).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert("RGB")

        # Create a unique filename using a hash of the image content
        filename = hashlib.sha1(image_content).hexdigest()[:10] + ".png"
        file_path = output_dir / filename

        # Save the image to the specified directory
        image.save(file_path, "PNG", quality=80)
        print(f"Image saved to {file_path}")
        return file_path
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_url}: {e}")
        return None
    except Exception as e:
        print(f"Error processing image {image_url}: {e}")
        return None

# Main function to get data, parse images, save URLs to CSV, and download images
def get_data(url, output_dir="images"):
    # Get the page content
    content = get_content_from_url(url)

    # Parse image URLs
    image_urls = parse_image_urls(content)

    # Save image URLs to CSV
    save_urls_to_csv(image_urls)

    # Download and save images to the specified directory
    for image_url in image_urls:
        get_and_save_image_to_file(image_url, output_dir, base_url=url)

# Example Usage
# url = "https://example.com"  # Replace this with your target URL
get_data(url, output_dir="/content/sample_data/CAP")  # Images will be saved in "images" folder
```

```python
model = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
feature_extractor = ViTImageProcessor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer = AutoTokenizer.from_pretrained("nlpconnect/vit-gpt2-image-captioning")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)



max_length = 16
num_beams = 4
gen_kwargs = {"max_length": max_length, "num_beams": num_beams}
def predict_step(folder_path):
    images = []
    image_paths = []

    # Load all images from the folder
    for filename in os.listdir(folder_path):
        if filename.endswith((".jpg", ".jpeg", ".png")):  # Add more image extensions if needed
            image_path = os.path.join(folder_path, filename)
            image_paths.append(image_path)
            i_image = Image.open(image_path)
            if i_image.mode != "RGB":
                i_image = i_image.convert(mode="RGB")

            i_image = i_image.resize((224, 224))  # Resize image to model's expected input size
            i_image = np.array(i_image).astype(np.float32) / 255.0  # Normalize image to [0, 1] range
            images.append(i_image)

    images = np.array(images)  # Convert list of images to a numpy array
    images = torch.tensor(images, dtype=torch.float32).permute(0, 3, 1, 2)  # Convert to tensor with shape (batch_size, channels, height, width)

    pixel_values = feature_extractor(images=images, return_tensors="pt", do_rescale=False).pixel_values
    pixel_values = pixel_values.to(device)

    output_ids = model.generate(pixel_values, **gen_kwargs)

    preds = tokenizer.batch_decode(output_ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)
    preds = [pred.strip() for pred in preds]

    # Combine all predictions into a single string with spaces between captions
    combined_caption = " ".join(preds)
    print(combined_caption)
    return combined_caption
```

```python
caption=predict_step("/content/sample_data/CAP")
```

```python
pipe = pipeline('summarization', model = 't5-large',max_length=120 ,truncation=True)
def split_text(text, max_chunk_length=512):
    # Split the text into chunks of max_chunk_length characters or smaller
    sentences = text.split('. ')
    current_chunk = []
    chunks = []

    for sentence in sentences:
        # If adding a sentence makes the chunk too long, create a new chunk
        if len(" ".join(current_chunk)) + len(sentence) <= max_chunk_length:
            current_chunk.append(sentence)
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentence]

    # Add the last chunk if it's not empty
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

def generate_summary(cleaned_text):
    # Split the text into chunks
    chunks = split_text(cleaned_text, max_chunk_length=512)

    # Generate summaries for each chunk
    summaries = []
    for chunk in chunks:
        summary = pipe(chunk)[0]['summary_text']
        summaries.append(summary)

    # Combine all the summaries into one big summary
    big_summary = " ".join(summaries)

    return big_summary

# Example usage
# cleaned_text = """Your very long text that exceeds the model's input length limit..."""
# big_summary = generate_summary(cleaned_text)
# big_summary = generate_summary()
```
